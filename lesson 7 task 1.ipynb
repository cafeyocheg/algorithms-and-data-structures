```python
# 1. Отсортируйте по убыванию методом пузырька одномерный целочисленный массив, 
# заданный случайными числами на промежутке [-100; 100). Выведите на экран исходный и отсортированный массивы.
# a. алгоритм сортировки должен быть в виде функции, которая принимает на вход массив данных,
# b. постарайтесь сделать алгоритм умнее, но помните, что у вас должна остаться сортировка пузырьком.
# Улучшенные версии сортировки, например, расчёской, шейкерная и другие в зачёт не идут.
# ----------------------------------------------------------------------------------------------------------------------------

import random

size = 25
array = [random.randint(-100, 99) for el in range(size)]
print(f'Исходный массив:\n{array}')

# метод первый: по сложности O(n2), как в видеоуроке, но с условием "по убыванию"
def bubble(a):
    n = 1
    while len(a) > n:
        for i in range(len(a) - 1):
            if a[i] < a[i + 1]:
                a[i], a[i + 1] = a[i + 1], a[i]
        n += 1    
    return array

print(f'Отсортированный массив по O(n2):\n{bubble(array)}')

# модификация (поменьше бы циклов): 
# сложность уменьшается до O(n) (?) - ускорился процесс всего в 1.5-2 раза
# при size=500:  time2/time1=2.12
# при size=100:  time2/time1=2.06
# при size=10:   time2/time1=2.66
def bubble_modified(a):
    for i in range(len(a) - 1, 0, -1):
        for j in range(i):
            if a[j] < a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
                
    return a
    
print(f'Отсортированный массив по O(n):\n{bubble_modified(array)}')

# -------------------------------------------------------------------------------------------------------------------
# проверка на улучшение

from timeit import timeit

executions = 1000
time1 = timeit(f'bubble({array})',
              setup='from __main__ import bubble',
              number=executions)
time2 = timeit(f'bubble_modified({array})',
              setup='from __main__ import bubble_modified',
              number=executions)
print(f'По времени выполнения модифицированный метод быстрее в {round(time1 / time2, 2)} раз')
```
